```python
#Exercice1:
import numpy as np
import matplotlib.pyplot as plt

# Paramtres
n = 1000           # Nombre dchantillons dans une exprience
repeats = 10       # Nombre de rptitions de lexprience

# Esprance thorique
expected_mean = 0.5

# Prparation de la figure
plt.figure(figsize=(10, 6))

# Rpter lexprience plusieurs fois
for i in range(repeats):
    # 1. Gnrer n ralisations de X ~ U(0, 1)
    X = np.random.uniform(0, 1, n)
    
    # 2. Calculer les moyennes cumules
    cumulative_means = np.cumsum(X) / np.arange(1, n + 1)
    
    # Tracer la moyenne cumule
    plt.plot(cumulative_means, label=f'Exprience {i+1}', alpha=0.6)

# Tracer la ligne de l'esprance thorique
plt.hlines(expected_mean, 0, n, colors='red', linestyles='dashed', label='Esprance thorique (0.5)')

# Mise en forme du graphique
plt.title('Convergence des moyennes cumules vers lesprance thorique de U(0,1)')
plt.xlabel('Nombre dchantillons')
plt.ylabel('Moyenne cumule')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Exercice3
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

# Paramtres
p = 0.3             # Probabilit de succs
n = 10000           # Nombre de simulations

# 1. Simulation de la variable gomtrique
# scipy.stats.geom retourne le nombre de tentatives avant le succs, donc X ~ Geom(p) donne k+1
# Pour avoir le nombre d'checs avant le succs, on fait -1
simulations = geom.rvs(p, size=n) - 1

# 2. Histogramme des frquences observes
max_k = np.max(simulations)
values, counts = np.unique(simulations, return_counts=True)
frequencies = counts / n

# 3. Distribution thorique
k_vals = np.arange(0, max_k + 1)
pmf_theorique = geom.pmf(k_vals + 1, p)  # +1 pour ajuster  la version scipy (nombre d'essais)

# Trac
plt.figure(figsize=(10, 6))
plt.bar(values, frequencies, width=0.8, alpha=0.6, label="Frquences observes")
plt.plot(k_vals, pmf_theorique, 'ro-', label="Distribution gomtrique thorique")

# Mise en forme
plt.title(f"Loi gomtrique : comparaison entre simulation et thorie (p = {p})")
plt.xlabel("Nombre dchecs avant le 1er succs")
plt.ylabel("Probabilit")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Exercice5
P = [[0.1, 0.6, 0.3],
     [0.4, 0.3, 0.3],
     [0.2, 0.5, 0.3]]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Matrice de transition
P = np.array([
    [0.1, 0.6, 0.3],
    [0.4, 0.3, 0.3],
    [0.2, 0.5, 0.3]
])

# Paramtres
n = 50            # Longueur de la trajectoire
N = 10000         # Nombre de trajectoires simules
nb_etats = P.shape[0]

# 2. Simulation de trajectoires
tats_finaux = []

for _ in range(N):
    etat = np.random.choice(nb_etats)  # tat initial alatoire
    for _ in range(n):
        etat = np.random.choice(nb_etats, p=P[etat])  # Transition
    tats_finaux.append(etat)

# 3. Estimation empirique
tats_finaux = np.array(tats_finaux)
probas_empiriques = np.bincount(tats_finaux, minlength=nb_etats) / N

# Affichage
tats = [f"tat {i}" for i in range(nb_etats)]

plt.figure(figsize=(8, 5))
plt.bar(tats, probas_empiriques, color='skyblue')
plt.title(f"Distribution empirique aprs {n} transitions (moyenne sur {N} chanes)")
plt.ylabel("Probabilit estime")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Affichage dans le terminal
for i, p in enumerate(probas_empiriques):
    print(f"Probabilit empirique dtre dans ltat {i} aprs {n} transitions : {p:.4f}")
```

```python
#Exercice7
import numpy as np

# Paramtres
mu = 5              # Moyenne
sigma = 2           # cart-type (donc variance = 4)
var = sigma**2
n = 100             # Nombre de variables par simulation
N = 10000           # Nombre de rptitions
epsilon = 0.5       # Seuil  pour l'vnement

# 1. Simulation des N moyennes empiriques de n variables iid
X = np.random.normal(loc=mu, scale=sigma, size=(N, n))
X_bar = X.mean(axis=1)

# 2. Calcul de la frquence empirique de lvnement |X - |  
ecarts = np.abs(X_bar - mu) >= epsilon
freq_empirique = np.mean(ecarts)

# 3. Borne thorique de Tchebychev
borne_tchebychev = var / (n * epsilon**2)

# Affichage
print(f"Frquence empirique : P(|X - |  )  {freq_empirique:.4f}")
print(f"Borne de Tchebychev : P(|X - |  )  {borne_tchebychev:.4f}")
```

```python
#exercice9
def aitken(u):
    """
    Applique la mthode d'Aitken   une suite u.
    u : tableau/list de rels (au moins de longueur 3)
    Retourne : liste contenant la suite acclre u'
    """
    u = np.array(u)
    n = len(u)
    u_prime = []

    for i in range(n - 2):
        delta1 = u[i+1] - u[i]
        delta2 = u[i+2] - 2*u[i+1] + u[i]
        if delta2 != 0:
            u_aitken = u[i] - (delta1**2) / delta2
            u_prime.append(u_aitken)
        else:
            u_prime.append(np.nan)  # ou u[i], selon prfrence

    return np.array(u_prime)
```

```python
#exercice11
import math

def convergents_sqrt2(n_terms):
    # Terme initial
    a = [1] + [2] * (n_terms - 1)

    # Initialisation des numrateurs et dnominateurs
    p = [a[0], a[0]*a[1] + 1]
    q = [1, a[1]]

    for n in range(2, n_terms):
        p_n = a[n] * p[n-1] + p[n-2]
        q_n = a[n] * q[n-1] + q[n-2]
        p.append(p_n)
        q.append(q_n)

    return list(zip(p, q))

# Comparaison avec math.sqrt(2)
n = 10
convergents = convergents_sqrt2(n)
sqrt2 = math.sqrt(2)

print(f"{'n':<3} {'p/q':<20} {'Approximation':<15} {'Erreur absolue'}")
for i, (p, q) in enumerate(convergents):
    approx = p / q
    error = abs(approx - sqrt2)
    print(f"{i:<3} {f'{p}/{q}':<20} {approx:<15.10f} {error:.2e}")
```

```python
#Exercice13
import numpy as np
import matplotlib.pyplot as plt

def suite_un(n_terms):
    u = [1.0]  # u = 1
    for _ in range(1, n_terms):
        u_next = u[-1] + 1 / u[-1]
        u.append(u_next)
    return np.array(u)

# Gnration
n = 100
u = suite_un(n)
n_vals = np.arange(n)

# Calcul de u_n^2 - 2n
asympt = u**2 - 2 * n_vals

# Tracs
plt.figure(figsize=(12, 5))

# u
plt.subplot(1, 2, 1)
plt.plot(n_vals, u, label='$u_n$', color='blue')
plt.title("volution de $u_n$")
plt.xlabel("n")
plt.ylabel("$u_n$")
plt.grid(True)

# u - 2n
plt.subplot(1, 2, 2)
plt.plot(n_vals, asympt, label="$u_n^2 - 2n$", color='red')
plt.axhline(y=1, color='gray', linestyle='--', label='Rfrence y=1')
plt.title("Comportement de $u_n^2 - 2n$")
plt.xlabel("n")
plt.ylabel("$u_n^2 - 2n$")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
